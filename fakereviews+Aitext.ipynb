## Top words

```python
from collections import Counter
import re

def get_top_words(text_series , n=20):
    words = ' '.join(text_series).lower()
    words = re.findall(r'\b[a-z]+\b',words)

    return Counter(words).most_common(n)
```

```python
print("Top words in genuine titles:", get_top_words(df_rev[df_rev['label']==0]['title'].fillna(''), n=15))
print("Top words in fake titles:", get_top_words(df_rev[df_rev['label']==1]['title'].fillna(''), n=15))
```

## Text Prepeocessing

### Removing \<br> tags

```python
df_rev['text_clean'] = df_rev['text'].str.replace(r'<br\s*/?>' ,' ', regex = True)
```

### Removing other html tags

```python
df_rev['text_clean'] = df_rev['text_clean'].str.replace(r'<[^>]+>' ,' ' , regex = True)
```

### Converting to lowercase

```python
df_rev['text_clean'] = df_rev['text_clean'].str.lower()
```

### Removing Punctuation marks

```python
df_rev['text_clean'] = df_rev['text_clean'].str.replace(r'[^\w\s]','',regex= True)
```

### Removing Extra Spaces

```python
df_rev['text_clean'] = df_rev['text_clean'].str.replace(r'\s+',' ' ,regex=True).str.strip()
```

## TRAIN TEST SPLIT

```python
df_rev.head()
```

```python
# full set of features - Model A
features_all = ['text_clean','rating','helpful_vote','verified_purchase','is_extreme_rating',
                'text_length','word_count','is_duplicate_text','title_length','is_duplicate_title']

# Reduced feature set- Model B (excluding the shorticut features)
features_no_shortcuts = ['text_clean', 'rating', 'helpful_vote', 'text_length', 'word_count', 
                          'is_extreme_rating', 'title_length']
```

# MODEL A (with shortcut features)

```python
## Train Test split
from sklearn.model_selection import train_test_split
X= df_rev[features_all]
Y = df_rev['label']
X_train, X_test , Y_train , Y_test = train_test_split(X,Y , test_size = 0.2 , random_state = 3 , stratify = Y)
```

```python
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
print("Train label balance:", Y_train.value_counts(normalize=True))
print("Test label balance:", Y_test.value_counts(normalize=True))
```

## TF-IDF Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
X_train_tfidf = tfidf.fit_transform(X_train['text_clean'])
X_test_tfidf = tfidf.transform(X_test['text_clean'])

print("TF-IDF train shape:", X_train_tfidf.shape)
print("TF-IDF test shape:", X_test_tfidf.shape)
```

## Standard Scaling and Combining features

```python
from scipy.sparse import hstack
from sklearn.preprocessing import StandardScaler
```

```python
numeric_cols = ['rating','helpful_vote','verified_purchase','is_extreme_rating',
                'text_length','word_count','is_duplicate_text','title_length','is_duplicate_title']

scaler = StandardScaler()

X_train_numeric = X_train[numeric_cols].astype(float).values
X_test_numeric = X_test[numeric_cols].astype(float).values

X_train_numeric_scaled = scaler.fit_transform(X_train_numeric)
X_test_numeric_scaled = scaler.transform(X_test_numeric)

X_train_final = hstack([X_train_tfidf,X_train_numeric_scaled])
X_test_final = hstack([X_test_tfidf, X_test_numeric_scaled])

print("Final train shape:", X_train_final.shape)
print("Final test shape:", X_test_final.shape)
```

## Logisitic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(max_iter = 1000 , random_state = 42)
log_reg.fit(X_train_final , Y_train)
print("Model trained successfully")
```

## Evaluation on Test Set

```python
from sklearn.metrics import classification_report , accuracy_score , confusion_matrix
```

### Predicted labels

```python
Y_pred = log_reg.predict(X_test_final)
```

```python
print("Accuracy:", accuracy_score(Y_test, Y_pred))
print("Classification Report: \n", classification_report(Y_test, Y_pred))
```

```python
## Confusion Matrix
cm = confusion_matrix(Y_test,Y_pred)
cm
```

```python
sns.heatmap(cm, annot = True , cmap = 'Reds' , xticklabels = ['Geniune' , 'Fake'] , yticklabels = ['Geniune','Fake'],fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Logistic Regression (Model A)')
```

## Observations:
- The model is clearly better at correctly classifying geniune reviews rather than identifying every fake review
- Currently , the model is predicting a decent amount of fake reviews as geniune but is good at identifying geniune reviews only a very less amount is predicted as fake.
- The model currently favours precision(out of all the predicted fakes how many are actually fake) over recall for fake labels.
- For geniune reviews recall(the amount of reviews actually predicted as geniune out of all the geniune reviews) is favoured.

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 100 , n_jobs = -1 , random_state = 42)
rf.fit(X_train_final,Y_train)
```

```python
Y_pred = rf.predict(X_test_final)
```

### Evaluation:

```python
print("RandomForest accuracy: " , accuracy_score(Y_test, Y_pred))
print("RF Classification Report: \n" , classification_report(Y_test, Y_pred))
```

```python
cm = confusion_matrix(Y_test,Y_pred)
sns.heatmap(cm , cmap = 'Greens' , annot = True , xticklabels =['geniune' , 'fake'] , yticklabels = ['geniune','fake'] , fmt ='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Random Forest (Model A)')
```

## Cross Validation

```python
from sklearn.model_selection import cross_val_score

# Logistic Regression CV
cv_scores_lr = cross_val_score(log_reg, X_train_final, Y_train, cv=5, scoring='accuracy', n_jobs=-1)
print("Logistic Regression CV scores:", cv_scores_lr)
print("Mean CV accuracy:", cv_scores_lr.mean())
print("Std dev:", cv_scores_lr.std())



# Random Forest CV
cv_scores_rf = cross_val_score(rf, X_train_final, Y_train, cv=5, scoring='accuracy', n_jobs=-1)
print("Random Forest CV scores:", cv_scores_rf)
print("Mean CV accuracy:", cv_scores_rf.mean())
print("Std dev:", cv_scores_rf.std())
```

## Observations:
- Both the models are quite stable across all the folds and hence are reliable(the accuracy isn't because of any lucky split).
- The random forest appears to be better than the logistic regression model in every aspect(precision/recall/f1score/acuracy).

## Loading the pre-trained gpt model and tokenizer

```python
from transformers import GPT2LMHeadModel , GPT2TokenizerFast
import torch

model_name = 'gpt2'
tokenizer = GPT2TokenizerFast.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
model.eval()  # since we aren't using the model for learning anything new and using it only for inference.

print("Model loaded successfully")
```

## Calculating Perplexity 

```python
def calculate_perplexity(text , tokenizer , model , max_length = 512):
    encodings = tokenizer(text , return_tensors = 'pt' , truncation = True , max_length = max_length)
    input_ids = encodings.input_ids

    with torch.no_grad():
        outputs = model(input_ids , labels = input_ids)
        loss = outputs.loss

    perplexity = torch.exp(loss).item()
    return perplexity
```

```python
sample_ai = df_ai[df_ai['label_ai_generated'] == 1]['text'].iloc[0]
sample_human = df_ai[df_ai['label_ai_generated'] == 0]['text'].iloc[0]

print("AI SAMPLE PERPLEXITY:" , calculate_perplexity(sample_ai , tokenizer , model))
print("Human SAMPLE PERPLEXITY:" , calculate_perplexity(sample_human , tokenizer , model))
```

```python
df_ai['perplexity'] = df_ai['text'].apply(lambda x: calculate_perplexity(x, tokenizer, model))

print("Mean perplexity - AI:", df_ai[df_ai['label_ai_generated']==1]['perplexity'].mean())
print("Mean perplexity - Human:", df_ai[df_ai['label_ai_generated']==0]['perplexity'].mean())
```

# Calculating Burstiness 
- How much the length of sentences varies with a single review text.
- expected : AI reviews- low burstiness and Human reviews- High burstiness

```python
def calculate_burstiness(text):
    sentences = text.split('.')
    sentences = [ s.strip() for s in sentences if s.strip() ]
    if len(sentences) < 2:
        return 0           ## 0 or 1 sentences only , have nothing to measure burstiness of.
    lengths = [len(s.split()) for s in sentences]
    burstiness = np.std(lengths)
    return burstiness
```

```python
df_ai['burstiness'] = df_ai['text'].apply(calculate_burstiness)
```

## Calculating Vocabulary diversity
- how much the words repeat within a review.
- high score = low repition and low score = more repition.

```python
def calculate_vocabulary_diversity(text):
    words = text.lower().split()
    if len(words) == 0:
        return 0
    return len(set(words))/ len(words)    
```

```python
df_ai['vocab_diversity'] = df_ai['text'].apply(calculate_vocabulary_diversity)
```

```python
df_ai.groupby('label_ai_generated')['vocab_diversity'].mean()
```

- As we have already observed that ai reviews are shorter in length (say 5 words) compared to human reviews there is comparatively less room for repitition as compared to a human revews (containing say 50 words) which results in a lower vocab_diversity for human reviews.
- It is a limitation of the metric and doesnt mean that human reviews are less diverse.
